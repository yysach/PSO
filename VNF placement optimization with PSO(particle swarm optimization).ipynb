```python
## reading the topology file
file = open("DataFiles/savvis.txt",'r')
content=file.readlines()
content
```

```python
## getting nodes and edges
cleaned = content[0].split('\n')[0].split(' ')
nodes = int(cleaned[0])
edges = int(cleaned[1])
print("nodes and edges are {} and  {}".format(nodes,edges))
```

```python
## need to initialized a matrix of size nodes*nodes
max_value = 9999999
adjacency_matrix= [ [ max_value for i in range(nodes) ] for j in range(nodes) ]
```

```python
## reading the content list
for i in range(1,len(content)):
    cleaned = content[i].split('\n')[0].split(' ')
    src = int(cleaned[0])
    dest = int(cleaned[1])
    latency = float(cleaned[2])
    adjacency_matrix[src][dest]=latency
    adjacency_matrix[dest][src]=latency

# applying floyd warshall shortest path algorithm for transitive property
for k in range(0,nodes):
    for i in range(0,nodes):
        for j in range(0,nodes):
            if adjacency_matrix[i][k]+adjacency_matrix[k][j]<adjacency_matrix[i][j]:
                # then update the matrix
                adjacency_matrix[i][j]=adjacency_matrix[i][k]+adjacency_matrix[k][j]

for i in range(0,nodes):
    adjacency_matrix[i][i]=max_value
adjacency_matrix
```

```python
# initialising load and resource vector 
import random
vnf = 4
load=[]
resource=[]

for i in range(0,vnf):
    load.append(random.randint(1,10))
    
for i in range(0,nodes):
    resource.append(random.randint(10,20))
    
print(load)
print(resource)
```

```python
# initialising bandwidth matrix (from source to destination)
bandwidth = [ [ 0 for i in range(nodes) ] for j in range(nodes) ]
for i in range(0,nodes):
    for j in range(i+1,nodes):
        x=random.randint(5,10)
        bandwidth[i][j]=x
        bandwidth[j][i]=x    
bandwidth
```

```python
# initializing traffic for vnfs
traffic = [ [ 0 for i in range(vnf) ] for j in range(vnf) ]
for i in range(0,vnf):
    for j in range(i+1,vnf):
        x=random.randint(1,5)
        traffic[i][j]=x
        traffic[j][i]=x    
traffic
```

```python
# function to check bandwidth constraint
# here param x is vnf sequence like [5,10,1,0]
def bandwidthConstraint(x):
    flag = True
    for i in range(0,vnf-1):
        flag = flag and traffic[i][i+1]<=bandwidth[x[i]][x[i+1]]    
    return flag
```

```python
## function to check load resource constraint
def loadConstraint(x):
    flag = True
    for i in range(0,vnf):
        flag = flag and load[i]<=resource[x[i]]    
    return flag
```

```python
## objective function
## param x is position vector
def objectiveFunction(x):
    total_latency=0
    position_tuple=[]
    for i in range(0,nodes):
        position_tuple.append(tuple([x[i],i]))  
        
    ## sorting (position,node) 
    position_tuple.sort(reverse=True)
    vnf_seq=[]
    
    for i  in range(0,vnf):
        vnf_seq.append(position_tuple[i][1])
    
    if(loadConstraint(vnf_seq)):
        if(bandwidthConstraint(vnf_seq)):
            for i in range(0,len(vnf_seq)-1):
                total_latency=total_latency+adjacency_matrix[vnf_seq[i]][vnf_seq[i+1]]
            print("vnf sequence: {} | Total Latency {}".format(vnf_seq,total_latency))
            return total_latency
        else:
            print("bandwidth error")
            return max_value
    else:
        print("load error")
        return max_value
```
